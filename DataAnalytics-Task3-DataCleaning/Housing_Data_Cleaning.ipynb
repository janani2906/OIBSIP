## Standardization

The categorical columns were checked for inconsistent values using the `unique()` function.

The values were already standardized (e.g., `yes`/`no`, `furnished`, `semi-furnished`, `unfurnished`), so no further standardization was required.

```python
Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df["price"] < lower) | (df["price"] > upper)]

print("Number of Outliers:", outliers.shape[0])
```

```python
df = df[(df["price"] >= lower) & (df["price"] <= upper)]
```

```python
print("Rows after removing outliers:", df.shape[0])
```

## Outlier Detection

The IQR (Interquartile Range) method was used to detect outliers in the **price** column.

Rows with values outside the lower and upper limits were removed to improve data quality.
